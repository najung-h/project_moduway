```python
import pandas as pd
import json

# 1. 데이터 로드
with open('../cleaned_data/kmooc_courses_public.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

df = pd.DataFrame(data)
```

```python
# 2. 강의명 + 교수자명 파생컬럼 content_group_key 생성
# 같은 커리큘럼인데 기수만 다른 경우 묶어서 처리하기 위함

# professor 결측치는 '미상'으로 대체
df['professor'] = df['professor'].fillna('미상')

# content_key 생성
df['content_key'] = df['name'] + "|" + df['professor']
```

```python
# 3. 요약(summary) 보강
# 요약이 너무 짧거나 비어있으면 강좌명으로 대체
df['summary'] = df['summary'].fillna("")
mask = df['summary'].str.len() < 10
df.loc[mask, 'summary'] = df.loc[mask, 'name']
```

```python
# 4. 불필요 컬럼 삭제
# 삭제할 컬럼 리스트 정의
drop_cols = ['detail_error_raw', 'public_yn', 'org', 'shortname']

# 1. 컬럼 삭제
df = df.drop(columns=drop_cols, errors='ignore')

# 2. 남길 컬럼들 순서 정리
ordered_cols = [
    'id', 'content_key', 'name', 'professor', 'org_name', 
    'classfy_name', 'middle_classfy_name', 'summary', 
    'url', 'course_image', 'enrollment_start', 'enrollment_end', 
    'study_start', 'study_end', 'week', 'course_playtime'
]
df = df[ordered_cols]
```

```python
# 저장
df.to_json('kmooc_processed_data.json', orient='records', force_ascii=False, indent=4)
df.to_csv('kmooc_processed_data.csv', index=False, encoding='utf-8-sig')
```
